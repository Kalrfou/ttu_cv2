```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
 
img = cv.imread('img_3.jpg')
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
grayscale_image = cv.cvtColor(img_rgb, cv.COLOR_BGR2GRAY)
plt.imshow(grayscale_image, cmap='gray')
```

```python
hist,bins = np.histogram(grayscale_image,256,[0,256])
plt.hist(hist,256,[0,256], color='r')
plt.show()
```

```python
ret, th = cv.threshold(grayscale_image, 90, 255, cv.THRESH_BINARY)
plt.imshow(th,cmap='gray')
```

```python
# Otsu's automatic thresholding

ret2, th2 = cv.threshold(grayscale_image, 0, 255,   cv.THRESH_BINARY + cv.THRESH_OTSU)
plt.imshow(th2, cmap='gray')
```

```python
# Adaptive thresholding
#Not really a good example!!!
thresh = cv.adaptiveThreshold(grayscale_image, 255,
                              cv.ADAPTIVE_THRESH_MEAN_C,
                              cv.THRESH_BINARY, 11, 2)
plt.imshow(thresh,cmap='gray')
```
