```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt
from skimage import io, color, filters, feature
```

```python
gray_cv = cv2.imread('Picture1.jpg')
img_rgb = cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB)
gray_sk = io.imread('Picture1.jpg')
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.imshow(img_rgb)
plt.title("OpenCV Image (RGB)")
plt.axis('off')
plt.subplot(1,2,2)
plt.imshow(img_sk)
plt.title("scikit-image Image (RGB)")
plt.axis('off') 
plt.show()
```

```python
blur_cv = cv2.GaussianBlur(gray_cv, (5,5), 0)
blur_sk = filters.gaussian(gray_sk, sigma=1)
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.imshow(blur_cv, cmap='gray')
plt.title("OpenCV Gaussian Blur")
plt.subplot(1,2,2)
plt.imshow(blur_sk, cmap='gray')
plt.title("scikit-image Gaussian Filter")
plt.show()
```

```python
edges_cv = cv2.Canny(gray_cv, 100, 200)
#edges_sk = feature.canny(gray_sk, sigma=1)
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.imshow(edges_cv, cmap='gray')
plt.title("OpenCV Canny Edges")
plt.subplot(1,2,2)
#plt.imshow(edges_sk, cmap='gray')
plt.title("scikit-image Canny Edges")
plt.show()
```
